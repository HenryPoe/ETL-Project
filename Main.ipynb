```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
income_csv = "datasets/ZipCode_Income/2019HouseholdIncome.csv"
income_df = pd.read_csv(income_csv)
income_df.head()
```

```python
carSales_csv = "datasets/used_car_sales.csv"
carSales_df = pd.read_csv(carSales_csv)
carSales_df.head()
```

```python
#drop unnecessary columns
carSales_df = carSales_df.drop(["ID"],axis=1)
carSales_df = carSales_df.drop(["DriveType"],axis=1)
carSales_df = carSales_df.drop(["BodyType"], axis=1)
carSales_df = carSales_df.drop(["Trim"], axis=1)
carSales_df = carSales_df.drop(["Engine"],axis=1)
carSales_df = carSales_df.drop(["NumCylinders"],axis=1)

carSales_df.head()
```

```python
#find count for cleaning
carSales_df.count()
```

```python
#rename columns and drop rows with missing values
clean_car_df = carSales_df.rename(columns={'pricesold':'Price_Sold','yearsold':'Year_Sold','zipcode':'Zip_Code'},inplace=False)
cleaned_df= clean_car_df.dropna()
cleaned_df.count()
```

```python
#bring zip code to the first column as we are basing our "analysis" on it
cleaned_df = cleaned_df[["Zip_Code","Price_Sold","Mileage","Make","Model","Year","Year_Sold"]]
cleaned_df.head()
```

```python
#filter out 2019 data for "comparison" with 2019 census data
cleaned_df_2019 = cleaned_df[cleaned_df["Year_Sold"] == 2019]
```

```python
#remove Year_Sold column
cleaned_df_2019 = cleaned_df_2019.drop(["Year_Sold"],axis=1)
```
